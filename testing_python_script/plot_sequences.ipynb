```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys
```

```python
ee_position_df = pd.read_csv(os.path.join('data_sim', 'positions_ee_backup.txt'), sep=',', header=0)
```

```python
ee_position_df
```

```python
plt.rcParams["axes.axisbelow"] = False
fig1, f1_ax = plt.subplots(ncols=1, nrows=4, figsize=(9,8), gridspec_kw={'height_ratios': [2, 2, 2, 1]})
fig1.subplots_adjust(hspace=0.5)

# x-axis
f1_ax[0].step(ee_position_df['time'], ee_position_df['pos_x'], c='r', label='EE desired x-position')
f1_ax[0].set_xlim([-3, 50])
f1_ax[0].set_ylabel('X-Axis\nPoisiton [m]\n(Task Space)')

# y-axis
f1_ax[1].step(ee_position_df['time'], ee_position_df['pos_y'], c='g', label='EE desired y-position')
f1_ax[1].set_xlim([-3, 50])
f1_ax[1].set_ylabel('Y-Axis\nPoisiton [m]\n(Task Space)')

# z-axis
f1_ax[2].step(ee_position_df['time'], ee_position_df['pos_z'], c='b', label='EE desired z-position')
f1_ax[2].set_xlim([-3, 50])
f1_ax[2].set_ylabel('Z-Axis\nPoisiton [m]\n(Task Space)')
f1_ax[2].set_xlabel('Time [s]')


fig1.align_ylabels(f1_ax[:])


prev_y = 0
for idx in range(0,6):
    f1_ax[3].barh(0, width=8, left=prev_y, color='w', edgecolor='k', height=2)
    f1_ax[3].text(prev_y+1, -0.25, 'Sequence '+str(ee_position_df.iloc[idx,-1]+1))
    f1_ax[3].set_xlim([-3, 50])
    
    f1_ax[3].axis('off')
    f1_ax[3].set_facecolor("none")
    prev_y += 8
    

plt.show()
```
